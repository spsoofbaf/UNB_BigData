Each record in the dataset represents a single taxi trip and includes information such as the start and end times and locations of the trip, the fare amount, and various other details about the trip and the taxi that was used.

Here are the fields included in each record of the taxi trip dataset:


*   **trip_id:** Unique identifier for each trip
*   **taxi_id:** Unique identifier for the taxi
*   **trip_start_timestamp:** Start time of the trip in UTC
*   **trip_end_timestamp:** End time of the trip in UTC
*   **trip_seconds:** Time duration of the trip in seconds
*   **trip_miles:** Distance of the trip in miles
*   **pickup_census_tract:** Census tract of the pickup location
*   **dropoff_census_tract:** Census tract of the dropoff location
*   **pickup_community_area:** Community area of the pickup location
*   **dropoff_community_area:** Community area of the dropoff location
*   **fare:** Total fare charged for the trip
*   **tips:** Total tip given by the passenger
*   **tolls:** Total tolls paid by the passenger
*   **extras:** Extra charges other than fare and tolls
*   **trip_total:** Total amount paid by the passenger including fare, tip, tolls, and extras
*   **payment_type:** Payment type used by the passenger (credit card, cash, etc.)
*   **company:** Taxi company that provided the service
*   **pickup_centroid_latitude:** Latitude of the pickup location
*   **pickup_centroid_longitude:** Longitude of the pickup location
*   **pickup_centroid_location:** Coordinates of the pickup location in the format of (latitude, longitude)
*   **dropoff_centroid_latitude:** Latitude of the dropoff location
*   **dropoff_centroid_longitude:** Longitude of the dropoff location
*   **dropoff_centroid_location:** Coordinates of the dropoff location in the format of (latitude, longitude)

5. Show how to filter data for download. For example a) datatype b) date/time range, c)
geographical region, d) ID (discuss queries).

To filter Taxi Trip data, we can append `$where` at the end of the url.

To filter by `datatype`, we can use different urls as previous question:

json_url = "https://data.cityofchicago.org/resource/wrvz-psew.json"

geojson_url = "https://data.cityofchicago.org/resource/wrvz-psew.geojson"

To filter by `date-time range` we can use `trip_start_timestamp`.

To filter by `location`, we can use these:

`pickup_centroid_latitude` and `pickup_centroid_longitude`

or

`dropoff_centroid_latitude` and `dropoff_centroid_longitude`


And to filter by `id`, we can pass the `trip_id` as a parameter to retrieve data.

````python
import requests

# a) Filtering by datatype:
print("\n \n ** a) Filtering by datatype:")

json_url = "https://data.cityofchicago.org/resource/wrvz-psew.json"
geojson_url = "https://data.cityofchicago.org/resource/wrvz-psew.geojson"

params = {
    "$limit": 2,
    "$where": "fare >= 5000",
    "$$app_token": 'pcIIzxS097oKu2Zq3lU1vZpsV'
}

response_json = requests.get(json_url, params=params)
data_json = response_json.json()

print("\n \n ** JSON Data: \n")
print(data_json)

response_geojson = requests.get(geojson_url, params=params)
data_geojson = response_geojson.json()

print("\n \n ** GeoJSON Data: \n")
print(data_geojson)

# b) Filtering by date/time range:

params = {
    "$limit": 2,
    "$where": "trip_start_timestamp >= '2022-01-01T00:00:00.000' AND trip_start_timestamp < '2022-02-01T00:00:00.000'",
    "$order": "trip_start_timestamp DESC",
    "$$app_token": 'pcIIzxS097oKu2Zq3lU1vZpsV'
}

response = requests.get(json_url, params=params)
data = response.json()

print("\n \n ** b) Filtering by date/time range: \n")
print(data)

# b) Filtering by location:

params = {
    "$limit": 2,
    "$where": "pickup_centroid_latitude > 41.8 AND pickup_centroid_latitude < 41.9 AND pickup_centroid_longitude > -87.7 AND pickup_centroid_longitude < -87.6",
    "$order": "trip_start_timestamp DESC",
    "$$app_token": 'pcIIzxS097oKu2Zq3lU1vZpsV'
}

response = requests.get(json_url, params=params)
data = response.json()

print("\n \n ** c) Filtering by location: \n")
print(data)

# b) Filtering by id:

params = {
    "$limit": 2,
    "$$app_token": 'pcIIzxS097oKu2Zq3lU1vZpsV'
}

response = requests.get(json_url + "?trip_id=0fbaf4500812630d1394c45ab198cc6afd5f5676", params=params)
data = response.json()

print("\n \n ** d) Filtering by id: \n")
print(data)
````
